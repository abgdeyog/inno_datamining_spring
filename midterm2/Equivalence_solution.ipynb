# Equivalence (10 - 25 pts)

### 1. Formulate definition of structural equvalence

Answer: Two vertices are structurally equivallent, if they have the same relationship to all other vertices

### 2. Formulate definition of automorphically equvalence


Answer: Two vertices u and v of a labeled graph G are automorphically equivalent if all the vertices can be re-labeled to form an isomorphic graph with the labels of u and v interchanged. Two automorphically equivalent vertices share exactly the same label-independent properties.

### 3. Generate two example datasets of each type

1. Generate datasets using `numpy`
2. Plot these three demonstrations using `pyplot` 


```python
#### 1 type demonstration code
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline

def draw_graph(N):
    adj_matrix = np.random.rand(N, N)
    adj_matrix = (adj_matrix + adj_matrix.T) / 2
    adj_matrix = np.int32(np.round(adj_matrix))
    inds = np.random.choice(N, 2, replace=False)

    for i in range(N):
        adj_matrix[i,i]=0
        
    
    adj_matrix[inds[0], :] = adj_matrix[inds[1], :]
    adj_matrix[:, inds[0]] = adj_matrix[:, inds[1]]
    
    print(adj_matrix)
    
    G = nx.from_numpy_array(adj_matrix)
    labels = ['green' if i in inds else 'red' for i in range(N)]
    nx.draw(G, node_color=labels, with_labels=True)
    plt.show()

draw_graph(10)
draw_graph(5)
```
